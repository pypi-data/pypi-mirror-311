```python
from nucdens import access
# create a test library locally using the standard web location for density files
densdf=access.database(workdir=os.environ["HOME"]+"/work/densdbtest")
```

```python
# group columns for selected printouts 
gridcols=['NP12A','NP12B','NP12','P12A','P12B', 'P12C',
          'NP34A','NP34B','NP34','P34A','P34B','P34C',
          'NP3A','NP3B','NP3','P3A', 'P3B', 'P3C',
           'NQ4A', 'NQ4B', 'NQ4', 'Q4A','Q4B','Q4C',
           'NQA','NQB','NQ','QA', 'QB','QC',
           'NPTA','NPTB', 'NPT', 'PTA', 'PTB', 'PTC',
           'NEA', 'NEB', 'NE','EA','EB', 'EC', 'NX']

pwcols=['taumax','lsummax','lmax','j12max','l3max','l4max','lammax','j12max3nf','j3max3nf','tau4max']
nuclcols=['kind','Z','N','E[MeV]', 'MTtot','Jtot']
NNcols=['MODENN','orderNN','LambdaNN','potnr','empotnr','relcalc']
TNFcols=['lam3NF','c1','c3','c4','cd','ce','tnforder']
TNFcolsN3LO=['cE1','cE2','cE3','cE4','cE5','cE6','cE7','cE8','cE9','cE10',]
SRGcols=['nnsrg','lambdaSRGNN', 'potbareNN','ostatSRGNN', 'cutnumSRGNN',]
SRG3Ncols=['tnfsrg','lambdaSRG3N','potbareNNTNF', 'ostatSRGTNF','cutnumSRGTNF', 'LambdaSRGTNF']
Runcols=['jubedirname','benchnr','runnr','hashname',
 'uniquefilename','localonly','localavail','addtime','moditime']

```

```python
# select some entries in the data base 
select=((densdf.pddf.Z==2)&(densdf.pddf.N==1)
      & (densdf.pddf.potnr==66.0) & (densdf.pddf.omega==131.857) 
      & (densdf.pddf.LambdaNN==1) & (densdf.pddf.orderNN=="nnlo")& (densdf.pddf.kind=="one")
      & (densdf.pddf.kind=="one")& (densdf.pddf.tnforder=="n2lo"))

#densdf.pddf.loc[select,"Jtot"]=1.0
densdf.pddf[select][nuclcols+NNcols+["Nmax"]+SRGcols+["omega","theta"]+["cE1","cE3","potnr"]+TNFcols].sort_values(by=["orderNN","LambdaNN","omega","theta"])
```
