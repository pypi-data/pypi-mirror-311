```python
# now we will fill the missing values
import pandas as pd
df=pd.read_csv('kidneyDisease.csv')
#Performing the EDA-EXPLORATORY DATA ANALYSIS
for column in df.columns:
    if df[column].dtype == 'object':
        df.fillna({column: df[column].mode()}, inplace=True)
    else:
        df.fillna({column: df[column].mean()}, inplace=True)

print(df.isnull().sum())
```

```python
# now we will handle the outliers
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
#Q1 first quartile or 25 % 
df=pd.read_csv('kidneyDisease.csv')
#Performing the EDA-EXPLORATORY DATA ANALYSIS
print('--------------------------------------------BEFORE-------------------------------')
sbn.boxplot(x=df['age'])
plt.show()
#we are now removing the outliers.
Quantile1=df['age'].quantile(0.25)#25% percentile
Quantile3=df['age'].quantile(0.75)#75 % percentile
IQR=Quantile3-Quantile1
#filtering our
df = df[~((df['age'] < (Quantile1 - 1.5 * IQR)) | (df['age'] > (Quantile3 + 1.5 * IQR)))]
print('-------------------------AFTER--------------------------------------------------------------')
sbn.boxplot(x=df['age'])
```

```python
#Validation checking
import pandas as pd
#loading the KIDNEY DISEASE DATASET
df=pd.read_csv('kidneyDisease.csv')
invalid_age = df[(df['age'] < 0) | (df['age'] > 100)]
if not invalid_age.empty:
    print("Invalid age entries found:\n", invalid_age)
# Validation check for blood pressure values
invalid_bp = df[(df['bp'] < 0) | (df['bp'] > 300)]
if not invalid_bp.empty:
    print("Invalid blood pressure entries found:\n", invalid_bp)
df.to_csv("preproccessed_kidney_disease.csv", index=False)
```
