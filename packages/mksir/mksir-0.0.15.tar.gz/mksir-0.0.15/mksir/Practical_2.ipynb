```python
for column in df.columns:
    if df[column].dtype=='object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)
```

```python
df=df.fillna(df.select_dtypes(include='number').mean())
```

```python
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
rainfallColumn=df.columns[2:]
```

```python
#applying standarization
scaler=StandardScaler()
df[rainfallColumn]=scaler.fit_transform(df[rainfallColumn])
df
```

```python
#normalizan
from sklearn.preprocessing import MinMaxScaler
normalizer=MinMaxScaler()
df[rainfallColumn]=normalizer.fit_transform(df[rainfallColumn])
```

```python
df
```

```python
#log 
import numpy as np
df[rainfallColumn]=np.log1p(df[rainfallColumn])
```

```python
df
```

```python
#aggregation of Data
rain_aggre=df.groupby(['SUBDIVISION']).mean().reset_index()
```

```python
rain_aggre
```

```python
#applying discretiztion
num_bins=5
df[rainfallColumn] = df[rainfallColumn].apply(lambda x: pd.qcut(x, num_bins, labels=False, duplicates='drop'))
```
